# Pass4Analysis

This is a guide on how to navigate the Pass 4 analysis already done 

First, log in to pa-pub and fiesta 

You will see a directory called J2019Pass4
Inside, there are more directories labeled:

    - DataJ2019Pass4
    - ModelJ2019Pass4  
    - PastAttempts 
    - ResidualsJ2019Pass4


Past Attempts contains all the files not used in the final analysis

## Data Files

The following directories are where you can find the data files are a part of the HAWC Collaboration's Pass 4 Data files.

maptree: /data/scratch/userspace/kmalone/datasets/energy/maps/combined/dataset-ch103-ch755/rejiggered-files/maptree-ch103-ch755.root

response file: /lustre/hawcz01/hawcroot/maps/GP-maps-startChunk103-stopChunk603/countsmaps/detRes-GP-allDecs.root

## Convert maptree and detector response file to .hdf5 files

This is the step that applies the model:

Use J2019_fit_logparabola.py to generate .hdf5 files for the model and the residual

Originally done on the local computer 

use scp to copy the .hdf5 files to fiesta

## .hdf5 files to fits files

Use script hal_hdf5_to_fits.py to convert .hdf5 files to fits files
--output adds a prefix to the fits files

(new_hal) [elainen@fiesta ModelandResidualsJ2019Pass4]$ python hal_hdf5_to_fits.py --input test_model.hdf5
